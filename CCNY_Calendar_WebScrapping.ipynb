```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

url='https://www.ccny.cuny.edu/registrar/fall'
r=requests.get(url)
html_doc=r.text
soup=BeautifulSoup(html_doc)
#print(soup.prettify())

table=soup.find_all('table')
#print(table)

info=[]
for td in table[0].find_all('td'):
    info.append(td.text)
print(info)

```

```python
rows=[]
for i in range (0, len(info), 3):
    rows.append({"Day": info[i], "Weekday": info[i+1], "Description": info[i+2]})

df=pd.DataFrame(rows)
df
```
