```python
import pandas as pandas
from bs4 import BeautifulSoup
import requests

url='https://www.ccny.cuny.edu/registrar/fall'
r=requests.get(url)
html_doc=r.text
soup=BeautifulSoup(html_doc)
#print(soup.prettify())

table=soup.find_all('table')
#print(table)

for td in table[0].find_all('td'):
    print(td.text)
```
